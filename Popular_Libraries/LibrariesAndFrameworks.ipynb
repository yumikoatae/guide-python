Python math Library
The math library provides a wide range of mathematical functions and constants. It is useful for performing mathematical operations that go beyond basic arithmetic.

Common Functions
math.sqrt(x): Returns the square root of x.

Example: math.sqrt(16) → 4.0
math.factorial(x): Returns the factorial of a non-negative integer x.

Example: math.factorial(5) → 120 (5!)
math.pow(x, y): Raises x to the power of y.

Example: math.pow(2, 3) → 8.0
math.sin(x), math.cos(x), math.tan(x): Returns the sine, cosine, and tangent of x (in radians).

Example: math.sin(math.pi / 2) → 1.0
math.log(x[, base]): Returns the logarithm of x to the specified base. If no base is specified, it returns the natural logarithm (base e).

Example: math.log(100, 10) → 2.0
math.pi: Constant for the value of π (pi).

Example: math.pi → 3.141592653589793
math.e: Constant for the value of Euler's number (e).

Example: math.e → 2.718281828459045

math.ceil(x): Rounds x up to the nearest integer.

Example: math.ceil(3.2) → 4
math.floor(x): Rounds x down to the nearest integer.

Example: math.floor(3.7) → 3

import time in Python
The time module in Python provides various time-related functions. It allows you to work with time and perform tasks like measuring time intervals, pausing execution, and formatting time.

Common Uses:
Sleep: Pause the execution of the program for a specified number of seconds.

python
Copiar código
import time
time.sleep(2)  # Pauses for 2 seconds
Current Time: Get the current time in seconds since the epoch (January 1, 1970).

python
Copiar código
current_time = time.time()
Formatting Time: Convert timestamps into a readable format using time.localtime() and time.strftime().

python
Copiar código
local_time = time.localtime()
formatted_time = time.strftime("%Y-%m-%d %H:%M:%S", local_time)
Measuring Time: Measure the execution time of code using time.perf_counter() for high-resolution timing.

python
Copiar código
start_time = time.perf_counter()
# Code to measure
elapsed_time = time.perf_counter() - start_time
Summary
Using import time, you can easily manage time-related tasks in your Python programs, from simple delays to detailed time measurements and formatting.

import random in Python
The random module in Python provides functions for generating random numbers and performing random operations. It’s widely used in scenarios such as simulations, games, and random selections.

Common Functions
random.random()

Returns a random floating-point number between 0.0 and 1.0.
python
Copiar código
import random
print(random.random())
random.randint(a, b)

Returns a random integer between a and b, inclusive.
python
Copiar código
print(random.randint(1, 10))  # Random integer between 1 and 10
random.choice(sequence)

Returns a randomly selected element from a non-empty sequence (like a list).
python
Copiar código
fruits = ['apple', 'banana', 'cherry']
print(random.choice(fruits))
random.shuffle(sequence)

Shuffles the elements of a list in place, randomly rearranging their order.
python
Copiar código
deck = [1, 2, 3, 4, 5]
random.shuffle(deck)
print(deck)  # Deck will be shuffled
random.sample(population, k)

Returns a list of k unique elements randomly chosen from the population sequence.
python
Copiar código
numbers = [1, 2, 3, 4, 5]
print(random.sample(numbers, 3))  # Randomly selects 3 unique numbers
random.uniform(a, b)

Returns a random floating-point number between a and b.
python
Copiar código
print(random.uniform(1.0, 10.0))
Summary
The random module is a powerful tool for generating random numbers and making random selections in Python. It offers various functions for different types of randomization, making it useful for a wide range of applications.

Modules in Python
A module in Python is a file that contains definitions of functions, classes, and variables that can be reused in other programs. Modules help organize code, promoting reuse and easier maintenance.

Creating a Module
To create a module, simply create a Python file with a .py extension. For example, a file named my_module.py can contain the following code:

python
Copiar código
# my_module.py

def greeting(name):
    return f"Hello, {name}!"

PI = 3.14159
Importing Modules
You can use the import keyword to include a module in your code. There are several ways to import modules:

Importing the Entire Module:

python
Copiar código
import my_module

print(my_module.greeting("Maria"))  # Output: "Hello, Maria!"
print(my_module.PI)                   # Output: 3.14159
Importing Specific Functions:

python
Copiar código
from my_module import greeting

print(greeting("João"))  # Output: "Hello, João!"
Importing with an Alias:

python
Copiar código
import my_module as mm

print(mm.greeting("Ana"))  # Output: "Hello, Ana!"
Standard Modules
Python comes with a standard library that includes many modules you can use without needing to install anything. Examples include:

math: Mathematical functions like sqrt(), sin(), etc.
os: Interaction with the operating system for file and directory manipulation.
sys: Access to variables and functions related to the Python interpreter.
Example with Standard Modules:

python
Copiar código
import math

root = math.sqrt(16)
print(root)  # Output: 4.0
Third-Party Modules
In addition to the standard library, there are many third-party modules that can be installed using package managers like pip. Examples include:

requests: For making HTTP requests easily.
numpy: For numerical computing and array manipulation.
Package Structure
Modules can be organized into packages, which are directories containing an __init__.py file and other modules. A package can be imported as a module, allowing for better organization of code.

Final Considerations
Modules are a fundamental part of programming in Python, enabling effective organization and reuse of code. They help keep code clean and modular, making collaboration and maintenance easier.
